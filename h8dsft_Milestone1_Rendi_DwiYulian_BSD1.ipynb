```python
df.drop_duplicates(subset=['Name'], inplace=True)
# Perintah menghapus salah satu data duplikat
```

- Dengan perintah di atas, data duplikat sudah dibersihkan dan tersisa total 962 baris data

- Selanjutnya saya akan merubah 'M' menjadi '000000' dan 'K' menjadi '000'
- Dengan cara mengalikan M degan 1jt dan K dengan 1rb sekaligus merubah tipe datanya menjadi float.

```python
#metode untuk merubah 'M' dan 'K' menjadi angka yang seharusnya
def numerik(value): #fungsi numerk
    if value.endswith('M'): #jika value berakhiran M
        return float(value.replace('M', '')) * 1000000 #replace menjadi float dan dikali 1jt
    elif value.endswith('K'): #jika value berakhiran K
        return float(value.replace('K', '')) * 1000 #replace menjadi float dan dikali 1rb
    else: # selain itu
        return float(value) #ubah data menjadi float

# Mengubah format angka pada kolom-kolom tertentu menjadi tipe data numerik
kolom_angka = ['Followers', 'Authentic Engagement', 'Engagement Avg.']#kolom yang ingin di ubah
for i in kolom_angka: #jika i didalam kolom angka yang ingin dirubah
    df[i] = df[i].apply(numerik) #terapkan kedalam fungsi numerik di atas
```

- Dengan Fungsi di atas, M , K dan tipe data pada kolom Followers, Authentic Engagement, Engagement Avg. sudah berubah dengan semestinya.

Berdasarkan Dataset Rank Top 1000 Instagram Followers Diatas. 
- Saya ingin mengetahui kategori apa yang paling banyak masuk kedalam jajaran top 1000 ini.

```python
# Menghitung jumlah akun Instagram dalam setiap kategori
category_counts = df['Category'].value_counts().head(10)

# Mengambil data untuk 3 kategori teratas
top_categories = category_counts.head(3)

# Membuat pie chart menggunakan Seaborn
plt.figure(figsize=(6, 6))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("plasma"))
plt.title('Top 3 Categories by Count')
plt.axis('equal')  # Membuat lingkaran menjadi lingkaran sempurna

# Menampilkan legenda
plt.legend(top_categories.index, loc="best")

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

- Terlihat bahwa kategori yang paling banyak dari dataset top 1000 instagram followers adalah kategori music.
- Hal ini menunjukan bahwa akun instagram dengan kategori music menjadi magnet daya tarik yang banyak di gemari dan diminati semua orang di seluruh dunia.

```python
# Filter data hanya untuk kategori "Music"
music_accounts = df[df['Category'] == 'Music']

# Mengurutkan akun dalam kategori "Music" berdasarkan jumlah followers
top_music_accounts = music_accounts.sort_values(by='Followers', ascending=False).head(10)

# Mengurutkan akun dalam kategori "Music" berdasarkan rata-rata Engagement Avg.
top_music_engagement = music_accounts.sort_values(by='Engagement Avg.', ascending=False).head(10)

# Mengurutkan akun dalam kategori "Music" berdasarkan Authentic Engagement
top_music_authentic_engagement = music_accounts.sort_values(by='Authentic Engagement', ascending=False).head(10)

# Mengatur gaya visual Seaborn
sns.set(style="whitegrid")

# Membuat subplot dengan 1 baris dan 3 kolom
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Grafik pertama: Top 10 Music Accounts by Followers
sns.barplot(x='Followers', y='Name', data=top_music_accounts, palette="CMRmap", ax=axes[0])
axes[0].set_title('Top 10 Music Accounts by Followers')
axes[0].set_xlabel('Number of Followers')
axes[0].set_ylabel('Account Name')

# Grafik ketiga: Top 10 Music Accounts by Authentic Engagement
sns.barplot(x='Authentic Engagement', y='Name', data=top_music_authentic_engagement, palette="viridis", ax=axes[2])
axes[1].set_title('Top 10 Music Accounts by Authentic Engagement')
axes[1].set_xlabel('Authentic Engagement')
axes[1].set_ylabel('Account Name')

# Grafik kedua: Top 10 Music Accounts by Engagement Avg.
sns.barplot(x='Engagement Avg.', y='Name', data=top_music_engagement, palette="seismic", ax=axes[1])
axes[2].set_title('Top 10 Music Accounts by Engagement Avg.')
axes[2].set_xlabel('Engagement Avg.')
axes[2].set_ylabel('Account Name')


# Menampilkan subplot
plt.tight_layout()
plt.show()
```

Selanjutnya saya ingin melakukan pengujian menggunakan two-sample independent test.
- Saya ingin menguji apakah jumlah authentic engagement yang dimiliki oleh akun Instagram musik berbeda dari jumlah authentic engagement yang dimiliki oleh akun Instagram Actors.

- H0 :

Tidak ada perbedaan yang signifikan antara rata-rata Authentic Engagement akun Instagram dalam kategori 'Music' dan 'Actors'

- H1 :

Terdapat perbedaan yang signifikan antara rata-rata Authentic Engagement akun Instagram dalam kategori 'Music' dan 'Actors'


```python
from scipy.stats import ttest_ind

# Filter data hanya untuk kategori "Music" dan "Actors"
music_accounts = df[df['Category'] == 'Music']
Actors_accounts = df[df['Category'] == 'Actors']

# Lakukan two-sample independent t-test
result = ttest_ind(music_accounts['Authentic Engagement'], Actors_accounts['Authentic Engagement'])

# Ambil nilai p-value dari hasil uji
p_value = result.pvalue
print("P-Value : ",p_value)

# Ambil nilai signifikansi (alpha)
alpha = 0.05

# Bandingkan p-value dengan alpha
if p_value < alpha:
    print("H1: Terdapat perbedaan yang signifikan antara rata-rata Authentic Engagement akun Instagram dalam kategori 'Music' dan 'Actors'.")
else:
    print("H0: Tidak ada perbedaan yang signifikan antara rata-rata Authentic Engagement akun Instagram dalam kategori 'Music' dan 'Actors'.")
```

#### Analisa:
- Karena p-value > 0.05
- Tidak cukup bukti untuk menolak H0
- Sehingga H0 diterima : Tidak ada perbedaan yang signifikan antara rata-rata Authentic Engagement akun Instagram dalam kategori 'Music' dan 'Actors'.
